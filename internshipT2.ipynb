```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/car_price_prediction_.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df['Car_Age'] = 2025 - df['Year']
```

```python
#Average Price by Brand
plt.figure()
df.groupby('Brand')['Price'].mean().plot(kind='bar')
plt.title("Average Car Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()
```

Insight: Tesla and BMW have higher average prices than Ford and Audi.

Why it matters: Brand strongly influences resale value.

Action: Premium brands should be priced higher even with similar specs.



```python
#Mileage vs Price
plt.figure()
plt.scatter(df['Mileage'], df['Price'])
plt.title("Car Price vs Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

Insight: Cars with higher mileage generally sell at lower prices.

Why it matters: Mileage is a major depreciation factor.

Action: Mileage should be weighted heavily in price estimation models.

```python
#Price by Fuel Type
plt.figure()
df.groupby('Fuel Type')['Price'].mean().plot(kind='bar')
plt.title("Average Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

Insight: Electric vehicles command higher average prices.

Why it matters: Market demand for EVs is increasing.

Action: Dealers should expect higher resale margins on EVs.

```python
#Car Age vs Price
plt.figure()
plt.scatter(df['Car_Age'], df['Price'])
plt.title("Car Price vs Car Age")
plt.xlabel("Car Age (Years)")
plt.ylabel("Price")
plt.show()
```

Insight: Older cars show clear price depreciation.

Why it matters: Age is a strong predictor of value loss.

Action: Pricing models must factor age non-linearly.

```python
#Condition vs Price
plt.figure()
df.groupby('Condition')['Price'].mean().plot(kind='bar')
plt.title("Average Price by Car Condition")
plt.xlabel("Condition")
plt.ylabel("Average Price")
plt.show()
```

Insight: New and Like New cars are priced significantly higher.

Why it matters: Condition directly impacts buyer willingness to pay.

Action: Accurate condition classification is critical.
